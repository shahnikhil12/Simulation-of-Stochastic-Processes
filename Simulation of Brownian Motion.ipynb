```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from matplotlib.pyplot import figure
import matplotlib.colors
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
```

```python
def dim1_BM(steps = 100, seed = 0):
    #Simulates a 1d Brownian Motion by generating then adding normally distributed increments
    np.random.seed(seed)
    increment = np.random.randn(steps)
    
    W = np.zeros(steps+1)
    
    for j in range(steps):
        W[j+1] = W[j] + increment[j]
    
    plt.plot(W)
    
```

```python
for j in range(10):
    dim1_BM(100000,j)
    plt.axhline(y=0, color='k', linestyle='-')
```

```python
def dim2_BM(steps = 50, seed = 0):
    #Simulates a 2d Brownian Motion by generating and adding normally distributed increments
    np.random.seed(seed)
    increments = np.random.randn(steps * 2)
    
    W_x = np.zeros(steps + 1)
    W_y = np.zeros(steps + 1)
    
    #Adds increments based on index number mod 2, with result indicating which vector to make next addition
    for j in range(steps * 2):
        if j % 2 == 0:
            k = int(j / 2)
            W_x[k + 1] = W_x[k] + increments[j]
            
        else:
            k = int(math.floor(j / 2))
            W_y[k + 1] = W_y[k] + increments[j]
        
    time = list(range(1, steps+2))
    
    
    #Plotting Information
    plt.scatter(W_x, W_y, c=time, cmap="inferno_r", s = 10 ** (3 - 0.75 * np.log10(steps)))
    plt.colorbar()
    plt.grid()
    colour_map = plt.cm.get_cmap("inferno_r")
    
    
    #For small number of steps, we overlay the scatter with a plot to interpolate between successive steps
    if steps <= 1000:
        for i in range(steps):
            seg_x = [W_x[i], W_x[i+1]]
            seg_y = [W_y[i], W_y[i+1]]       
            colour = colour_map(time[i]/steps)
            plt.plot(seg_x,seg_y, c = colour, linewidth = 1)
    else:
        pass
    
    
    
```

```python
dim2_BM(10000,0)
```

```python
def dim3_BM(steps = 100, seed = 0):
    #Generates a 3d Brownian Motion by generating and adding normally distributed increments
    np.random.seed(seed)
    increments = np.random.randn(steps * 3)
    
    W_x = np.zeros(steps + 1)
    W_y = np.zeros(steps + 1)
    W_z = np.zeros(steps + 1)
    
    #Adds increments to each component based on mod 3 result of index value
    for j in range(steps * 2):
        if j % 3 == 0:
            k = int(j / 3)
            W_x[k + 1] = W_x[k] + increments[j]
            
        elif j % 3 == 1:
            k = int(math.floor(j / 3))
            W_y[k + 1] = W_y[k] + increments[j]
            
        elif j % 3 == 2:
            k = int(math.floor(j / 3))
            W_z[k + 1] = W_z[k] + increments[j] 
    
    time = list(range(1, steps+2))
    
    
    #Plotting information
    #Could use some more work to make perspective clearer
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    
    ax.scatter(W_x, W_y, W_z, c=time, cmap="inferno_r", s = 10 ** (3 - 0.75 * np.log10(steps)))
    
    colour_map = plt.cm.get_cmap("inferno_r")
    
    plt.grid() 
    plt.show()
    
   
```

```python
dim3_BM(100000)
```
