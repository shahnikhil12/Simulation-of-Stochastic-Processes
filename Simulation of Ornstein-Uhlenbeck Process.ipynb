```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from matplotlib.pyplot import figure
import matplotlib.colors

%matplotlib inline
```

```python
def ornstein_uhlenbeck(steps = 100, seed = 0, theta = 1, mu = 0, sigma = 1, dt = 0.01):
    np.random.seed(seed)
    x = np.zeros(steps + 1)
    
    for i in range(steps):
        x[i+1] = x[i] + theta * (mu - x[i]) * dt + sigma * float(np.sqrt(dt)) * np.random.randn()
        
    plt.plot(x)
```

```python
ornstein_uhlenbeck(steps = 10000, seed = 0, theta = 1, mu = 5, sigma = 1, dt = 0.01)
```

```python
for i in [0, 5, 10, 20, 50]:
    ornstein_uhlenbeck(steps = 100, seed = i, theta = 1, mu = 0, sigma = 1, dt = 2)
```
